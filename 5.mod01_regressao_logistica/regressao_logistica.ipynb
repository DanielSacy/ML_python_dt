```python
data_process = data
missing_percentual = (data_process.isnull().sum() / len(data_process['Product.ID'])) * 100
print(missing_percentual)
```

```python
# columns_to_drop = ['q8.2', 'q8.8', 'q8.9', 'q8.10', 'q8.17', 'q8.18','q8.20']
data_process = data_process.drop(['q8.2', 'q8.8', 'q8.9', 'q8.10', 'q8.17', 'q8.18','q8.20'], axis=1)
data_process.shape
```

```python
data_process = data_process.drop(['Product','Respondent.ID', 'q1_1.personal.opinion.of.this.Deodorant'], axis=1)
# data_process = data_process.drop(['Product','Respondent.ID'], axis=1)
data_process.shape
```

```python
data_process['q8.12'].fillna(data_process['q8.12'].median(), inplace=True)
data_process['q8.7'].fillna(data_process['q8.7'].median(), inplace=True)
```

```python
missing_percentual = (data_process.isnull().sum() / len(data_process['Product.ID'])) * 100
print(missing_percentual)
```

```python
label = data_process['Instant.Liking']
features = data_process.drop(['Instant.Liking'], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

#Setting the folds
SKfold = StratifiedKFold(n_splits=5)

#Setting the model
model = LogisticRegression(penalty='l2', solver='liblinear')
result = cross_val_score(model, features, label, cv=SKfold)

print(f'{result.mean():.15f}')
```

```python
#LOGISTIC REGRESSION ALTERING PARAMETERS
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

#Setting the values test
C_values = np.array([0.01,0.1,0.5,1,2,3,5,10,20,50,100])
regularization = ['l1','l2']
grid_values = {'C': C_values, 'penalty': regularization}

#Setting the model
model = LogisticRegression(penalty='l2', solver='liblinear')

#Grids
logistic_regression_grid = GridSearchCV(model, grid_values, cv=5)
logistic_regression_grid.fit(features,label)

#Printing best results
print(f'Best Accuracy {logistic_regression_grid.best_score_:^50}')
print(f'Best C Parameter {logistic_regression_grid.best_estimator_.C:^50}')
print(f'Best Penalty {logistic_regression_grid.best_estimator_.penalty:^50}')
```
