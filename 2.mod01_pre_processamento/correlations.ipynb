```python
import pandas as pd
pd.set_option("display.max_columns", None)
data = pd.read_csv("E:/DAY2DAY/Python_MachineLearning_DidaticaTech/ML_python_dt/2.mod01_pre_processamento/dataset/2015-building-energy-benchmarking.csv")
```

```python
data.dtypes
```

```python
data_int = data.select_dtypes(include=['int64', 'float64'])
```

```python
data_int.dtypes
```

```python
pd.set_option('display.width', 10)
correlations = data_int.corr(method='pearson')
print(correlations)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

plt.figure(figsize=(20,20))
heatmap = sns.heatmap(correlations)
heatmap.tick_params(labelsize=20)
plt.show()
```
